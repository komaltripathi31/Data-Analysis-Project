```python
df.replace("?",np.nan,inplace=True)
df.head(5)
```

```python
#Evaluating for missing data
```

```python
miss_data=df.isnull()
miss_data.head(5)
```

```python
df.isnull().sum()
```

```python
#According to above analysis,we can see that symboling contains zero null values,normalized-losses contain 41 null values and bore contains 4 null values
```

```python
#For dealing with these missing values we cannot drop whole column beacaue it might cause losing data 
#So we can replace these missing values by the average of all the values
```

```python
columns=["normalized-losses","bore","stroke","horsepower","peak-rpm"]
for column in columns:
    avg=df[column].astype("float").mean(axis=0)
    df[column].replace(np.nan,avg,inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
#we can see that the four doors are the most common type.
#so we can replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.dropna(subset=["price"], axis=0, inplace= True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
#Now,We obtain the dataset with no missing values
```

```python
df.dtypes
```

```python
#As we can see that many columns do not have correct data types like price should be float but it cannot and so on.
```

```python
#Making suitable data types of columns
```

```python
df[["bore","stroke"]]=df[["bore","stroke"]].astype("float")
df[["normalized-losses"]]=df[["normalized-losses"]].astype("int")
df[["price"]]=df[["price"]].astype("float")
df[["peak-rpm"]]=df[["peak-rpm"]].astype("float")
```

```python
#After conversion of data types
df.dtypes
```

```python
#As we can see that fuel consumption columns('city-mpg','highway-mpg')are extracting data in miles per gallon but if we change this unit into L/100km.
```

```python
df['city-L/100km']=235/df["city-mpg"]
df["highway-L/100km"]=235/df["highway-mpg"]
df[['city-L/100km',"city-mpg","highway-L/100km","highway-mpg"]].head()
```

```python
#Binning process
#As we can see that we could not identify which vehicle has high horsepower and which one has low
```

```python
df["horsepower"]=df["horsepower"].astype(float,copy=True)
```

```python
binwidth=(max(df["horsepower"])-min(df["horsepower"]))/4
binwidth
```

```python
#so we have created a bin array with a minimum value to a maximum value
bins=np.arange(min(df["horsepower"]),max(df["horsepower"]),binwidth)
bins
```

```python
#creating group names
group_names=['Low','Medium','High']
```

```python
df['horsepower-binned']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Anova analysis(analysis of variance)
#It analyzes the difference between different groups of the same variable.
```

```python
df_group = df[['drive-wheels','price']]
groups = df_group[['drive-wheels','price']].groupby(['drive-wheels'])
```

```python
groups.head(2)
```

```python
groups.get_group('4wd')['price']
```

```python
f_val, p_val = stats.f_oneway(groups.get_group('fwd')['price'], groups.get_group('rwd')['price'], groups.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
#Large F score means these three variables are highly correlated with price
```

```python
#Examine These three variables seperately
f_val, p_val = stats.f_oneway(groups.get_group('fwd')['price'], groups.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
f_val, p_val = stats.f_oneway(groups.get_group('4wd')['price'], groups.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
f_val, p_val = stats.f_oneway(groups.get_group('4wd')['price'], groups.get_group('fwd')['price'])  
 
print("ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
#It describes that in drive-wheels , the group contains fwd and rwd are most correlated with price
```

```python
#Model development
from sklearn.linear_model import LinearRegression
```

```python
df_new = df[['horsepower','curb-weight','engine-size','highway-mpg','bore','wheel-base','city-mpg','length','width']]
```

```python
lm=LinearRegression()
lm
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_new,df['price'],test_size=0.20, random_state=1)
print("The number of test samples: ",x_test.shape[0])
print("The number of train samples: ",x_train.shape[0])
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
print("The R_squared value for Multiple Linear Regression Model is: ",lm.score(x_test,y_test))
```

```python
predicted = lm.predict(x_test)
plt.figure(figsize=(12,4))
ax1=sns.distplot(df['price'],hist=False,color="r",label="Actual Values")
sns.distplot(predicted,hist=False,color="b",label="Predicted Values",ax=ax1)

plt.title("Actual vs Predicted Values for Price")
plt.xlabel("Price (in dollars)")
plt.ylabel("Proportion of Cars")

plt.show()
plt.close()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
pipe=Pipeline(Input)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
print("The R_squared value for Polynomial Linear Regression Model is: ",pipe.score(x_test,y_test))
```

```python
predicted = pipe.predict(x_test)
plt.figure(figsize=(12,4))
ax1=sns.distplot(df['price'],hist=False,color="r",label="Actual Values")
sns.distplot(predicted,hist=False,color="b",label="Predicted Values",ax=ax1)

plt.title("Actual vs Predicted Values for Price")
plt.xlabel("Price (in dollars)")
plt.ylabel("Proportion of Cars")

plt.show()
plt.close()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf = RandomForestRegressor()
Rf.fit(x_train,y_train)
```

```python
print("The R_squared value for Polynomial Linear Regression Model is: ",Rf.score(x_test,y_test))
```

```python
predicted = Rf.predict(x_test)
plt.figure(figsize=(12,4))
ax1=sns.distplot(df['price'],hist=False,color="r",label="Actual Values")
sns.distplot(predicted,hist=False,color="b",label="Predicted Values",ax=ax1)

plt.title("Actual vs Predicted Values for Price")
plt.xlabel("Price (in dollars)")
plt.ylabel("Proportion of Cars")

plt.show()
plt.close()
```
